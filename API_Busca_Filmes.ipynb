```python
import requests
import json
from IPython.core.display import Image, display

def requisicao(titulo):
    try:
        req = requests.get('http://www.omdbapi.com/?t=' + titulo + '&apikey=aabc817d')
        dicionario = json.loads(req.text)
        return dicionario
    except:
        print('Erro na conexão')
        return None

def imprimir_detalhes(filme):
    try:
        #r = requests.get(filme['Poster'])
        
        #with open('img.jpg', 'wb') as f:
        #    f.write(r.content)
            
        print('Título:', filme['Title'])
        print('Ano:', filme['Year'])
        print('Diretor:', filme['Director'])
        print('Atores:', filme['Actors'])
        print('Nota:', filme['imdbRating'])
        print('Premio:', filme['Awards'])
        print('Poster:', filme['Poster'])
        url = filme['Poster']
        
        if url == 'N/A':
            print('Este filme não possui poster!')
        else:
            
            display(Image(url, width=200, unconfined=True))
        print('')
    except:
        print('Erro na conexão')
        return None
    
sair = False
while not sair:
    op = input('Escreve o nome de um filme ou SAIR para finalizar o programa:')
    
    if op == 'SAIR':
        sair = True
        print('Saindo do programa...')
    else:
        filme = requisicao(op)
        if filme['Response'] == 'False':
            print('Filme não encontrado!')
        else:
            imprimir_detalhes(filme)
```
